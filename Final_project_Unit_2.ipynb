### Research Question 1: What census data is most associated total guns per capita?

```python
# 2) Which states have had the highest growth in total guns registrations per capita?
df_combined['high_gun_regist_x_capita_2010'] = df_combined["totals_2010"] / df_combined["Population_estimates_2010"] 
df_combined['high_gun_regist_x_capita_2016'] = df_combined["totals_2016"] / df_combined["Population_estimates_2016"] 
df_combined['dif_regist_high_gun_x_capita'] = df_combined['high_gun_regist_x_capita_2016'] - df_combined['high_gun_regist_x_capita_2010']
df_combined['inc_regist_high_gun_x_capita'] = df_combined['dif_regist_high_gun_x_capita'] / df_combined['high_gun_regist_x_capita_2010']
df_combined['inc_regist_high_gun_x_capita_%'] = df_combined['inc_regist_high_gun_x_capita'] * 100
df_combined.groupby('state')['inc_regist_high_gun_x_capita_%'].sum()
```

#### We can see that the census data most associated with total guns per capita is Indiana twice as far apart from the second.

```python
df_combined.groupby('state').Population_estimates_2016.sum()
```

#### We can also see that Indiana is by no means the country with the most population, so the amount of population is not necessarily related to the number of total guns.

### Research Question 2: What country have most handgun per capita in the year 2010?

```python
# 3) Which states has the highest average number of handguns per capita in the year 2010? 
df_combined_2 = df_combined.copy()
df_combined_2['totals_handguns_2010'] = df_combined_2['handgun_2010']+df_combined_2['redemption_handgun_2010']+df_combined_2['returned_handgun_2010']+df_combined_2['rentals_handgun_2010']+df_combined_2['private_sale_handgun_2010']+df_combined_2['return_to_seller_handgun_2010']
df_combined_2['handgun_x_capita_2010'] = df_combined_2["totals_handguns_2010"] / df_combined_2["Population_estimates_2010"] 
df_combined_2['handgun_x_capita_2010_%'] = df_combined_2['handgun_x_capita_2010'] * 100
df_combined_2.groupby('state')['handgun_x_capita_2010_%'].sum()
```

#### The country with most handguns registrations in 2010 is Alaska with 0,33%.

### Research Question 3: Which states have had the high number of permits per capita?

```python
# 2) Which states have had the most handguns registrations per capita in 2016?
df_combined_3 = df_combined.copy()
df_combined_3['totals_handguns_2016'] = df_combined_3['handgun_2016']+df_combined_3['redemption_handgun_2016']+df_combined_3['returned_handgun_2016']+df_combined_3['rentals_handgun_2016']+df_combined_3['private_sale_handgun_2016']+df_combined_3['return_to_seller_handgun_2016']
df_combined_3['handgun_x_capita_2016'] = df_combined_3["totals_handguns_2016"] / df_combined_3["Population_estimates_2016"] 
df_combined_3['handgun_x_capita_2016_%'] = df_combined_3['handgun_x_capita_2016'] * 100
df_combined_3.groupby('state')['handgun_x_capita_2016_%'].sum()
```

#### The country with most handguns registrations in 2016 is West Virginia with 0,48%.

### Research Question 4. What is the overall trend of high guns purchases?

```python
# 4) What is the overall trend of handgun purchases?
#df_combined.plot(x='Population_estimates_2016', y='totals_2010', kind='line');

trend_high_gun_df = gun_df_v1.copy()
trend_high_gun_df['total_high_guns'] = gun_df_v1['handgun']+gun_df_v1['redemption_handgun']+gun_df_v1['returned_handgun']+gun_df_v1['rentals_handgun']+gun_df_v1['private_sale_handgun']+gun_df_v1['return_to_seller_handgun']
trend_high_gun_df_v2 = trend_high_gun_df.groupby('month')['total_high_guns'].sum()

trend_high_gun_df_v2.plot(x='monh', y='total_high_guns', kind='bar', figsize=(300,100), fontsize=250, subplots=True, sharex=True);
```
